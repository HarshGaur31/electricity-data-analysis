```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("electricity.csv")

df.head()
```

```python
df.info()
```

```python
average_price = df["nswprice"].mean()
print(average_price)
```

```python
df.groupby("day")["nswprice"].mean().plot(kind="bar")
plt.xlabel("day")
plt.ylabel("average nsw price")
plt.title("average nsw price by day")
plt.show()
```

```python
plt.scatter(df["nswdemand"],df["nswprice"])
plt.xlabel("nsw demand")
plt.ylabel("nsw price")
plt.title("nsw demand vs price")
plt.show()
```

```python
corr = df.select_dtypes(include='number').corr()

import matplotlib.pyplot as plt

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("correlation heatmap")
plt.show()
```
